<h3>Importing all the necessary libraries</h3>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('./zomato.csv', encoding='latin-1')
df.info()
```

<h3>Scatter Plot</h3>

```python
x = np.random.random(50)*100
y = np.random.random(50)*100

plt.scatter(x, y,marker="*",c='c')
plt.show()
```

```python
sns.scatterplot(x=df["Aggregate rating"], y=df["Votes"], hue=df["Price range"], style=df["Has Online delivery"], size=df["Average Cost for two"])
plt.show()
```

<h3>Line Chart</h3>

```python
year = [2015 + x for x in range(10)]
weight = [78, 79, 80, 79, 76, 83, 84, 75, 78, 87]

plt.plot(year, weight, c="#ff3399", linestyle='--', lw=2)
plt.show()
```

<h3>Bar Chart</h3>

```python
year = [2015 + x for x in range(10)]
votes = np.random.randint(0, 10000, 10)

plt.bar(year, votes, color='c', edgecolor='black', align='center', width=0.8)
plt.show()
```

```python
sns.barplot(x=df["Price range"], y=df["Aggregate rating"])
plt.show()
```

<h3>Histrogram<h3>

```python
ages = np.random.normal(50, 1.2, 1000)


plt.hist(ages, bins=30, color='c', edgecolor='black')
plt.show()
```

<h3>Pie Chart</h3>

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 25, 39, 12]

plt.pie(students, labels=langs, colors=['c', 'm', 'r', 'b', 'g'], shadow=True, autopct='%1.1f%%', explode=(0, 0, 0.1, 0.11, 0), startangle=90)
plt.show()
```

<h3>Boxplot</h3>

```python
rnd = [10, 11, 12, 11, 11, 12, 13, 14, 14, 17, 15, 22]

plt.boxplot(rnd, vert=False, patch_artist=True)
plt.show()
```

```python
sns.boxplot(df["Price range"])
plt.show()
```
